```python
class Hashmap:
  def __init__(self, cap):
    self.cap = cap
    self.size = 0
    self.bucket = [ [] for _ in range(cap)]

  def __str__(self):
    return str(self.bucket)

  def __contains__(self, key):
    index = self._hash_func(key)
    bucket = self.bucket[index]

    for k,v in bucket:
      if k == key:
        return True
    return False

  def __len__(self):
    return self.size

  def add(self, key, value):
    index = self._hash_func(key)
    bucket = self.bucket[index]

    for i, (k,v) in enumerate(bucket):
      if k ==key:
        bucket[i] = (key, value)
        break
    else:
      bucket.append((key,value))
      self.size+=1

  def get(self, key):

    index = self._hash_func(key)
    bucket = self.bucket[index]
    for k,v in bucket:
      if k == key:
        return v
    raise KeyError("key not found")
  def remove(self, key):
    index = self._hash_func(key)
    bucket = self.bucket[index]

    for i, (k, v) in enumerate(bucket):
      if k == key:
        del bucket[i]
        self.size -= 1
        break
    else:
      raise KeyError(" Key not present")

  def keys(self):
    return [k for bucket in self.bucket for k, _ in bucket]
  def vals (self):
     return [v for bucket in self.bucket for _, v in bucket]

  def _hash_func(self, key):

    key_string = str(key)

    sum_of_vals = 0
    for i in key_string:
      sum_of_vals += ord(i)
    return sum_of_vals % self.cap

N = Hashmap(5)
N.add("key","nyanza")
N.add("key1","nyanza1")
N.add("key2","nyanza2")
print(N)
```
