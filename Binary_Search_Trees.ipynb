```python
class Node:
  def __init__(self, key):
    self.key = key
    self.left = None
    self.right = None
    self.value = None
    self.parent = None

  def __rpr__(self):
    return f"({self.key} , {self.value})"


class BinarySearchTree:
  def __init__(self):
    self.root = None

  #o(n) worst case time complexity, o(log n) average case

  def __contains__(self, key):

    currentN = self.root
    while currentN:
      if key < currentN.key:
        currentN = currentN.left
      elif key > currentN.key:
        currentN = currentN
      else :
        return True
    return False


  # o(n)
  def __itr__(self):

    yield from self._inorder(self.root)

  def __rpr__(self):
    return str(list(self._inorder(self.root)))

  def __str__(self):
    return str(list(self._inorder(self.root)))

  # o(n) worst case time complexity, o(log n) average case

  def insert(self, key, value):

    '''inserting the first ever node in the tree'''
    if self.root is None:
      self.root = Node(key)
      self.root.value = value

    else:

      '''updating the value with the given key'''
      currentN = self.root
      while True:
        if key < currentN.key:
          if currentN.left is None:
            currentN.left = Node(key)
            currentN.left.value = value
            currentN.left.pareent = currentN
            break
          else:
            currentN = currentN.left
        elif key > currentN.key:
          if currentN.right is None:
            currentN.right = Node(key)
            currentN.right.value = value
            currentN.right.parent = currentN
            break
          else:
            currentN = currentN.right
        else:
          currentN.value = value # updates if the value for the key we are looking for is the given key
          break


  def search(self, key):

    '''if the current Node is None then we have no value and should return none
      if the value is in the current node then should return the current node '''
    currentN = self.root

    while True:
      if currentN is None or currentN.key == key:
        return currentN

        '''if the key is in the next level of node and is less than the root, then we should
        check the left side for the value. '''
      elif key < currentN.key:
        if currentN.left is None: # If the value is not in the left side return None
          return None
        else:
          currentN = currentN.left
      else:
        if currentN.right is None: # If the value is not in the right side return None
          return None
        else:
          currentN = currentN.right


  # o(n) worst case time complexity, o(log n) average case

  def delete(self, key):
    # search for the node first
    node = self.search(key)
    if node is not None:
      self._delete(node)

    else:
      raise KeyError("Key not found in the tree")

  #o(n)

  def traverse(self, order):

    if order == "preorder":
      yield from self._preorder(self.root)

    elif order == "inorder":
      yield from self._inorder(self.root)

    elif order == "postorder":
      yield from self._postorder(self.root)

    else:
      raise ValueError("Invalid order")

  def _delete(self, node):
    # deleting a leaf node
    if node.right is None and node.left is None:
      if node.parent is None:
        self.root = None

      else:
        if node.parent.right == node:
          node.parent.right = None
        else:
          node.parent.left = None
        node.parent = None

    # deleting parent with one child node

    elif node.right is None or node.left is None:
      childN =node.right if node.right is not None else node.left

      if node.parent is None:
        childN.parent = None
        self.root = childN
      else:
        if node.parent.right == node:
          node.parent.right = childN
        else:
          node.parent.left = childN
        childN.parent = node.parent

      node.parent = node.right = node.left = None

    # deleting parent with two child nodes
    else:
      successor = self._successor(node)

      node.key = successor.key
      node.value = successor.value
      self._delete(successor)

  def _successor(self, node):
    if node is None:
      return ValueError(" There's no successor of None ")

    if node.right is None:
      return None

    else:
      currentN = node.right
      while currentN.left:
        currentN = currentN.left
      return currentN

  def _predecessor(self, node):
    if node is None:
      return ValueError(" There's no predecessor of None ")

    if node.left is None:
      return None

    else:
      currentN = node.left
      while currentN.right:
        currentN = currentN.right
      return currentN

  def _preorder(self, node):
    if node is not None:
      yield (node.key, node.value)
      yield from self._preorder(node.left)
      yield from self._preorder(node.right)

  def _inorder(self, node):
    if node is not None:
      yield from self._inorder(node.left)
      yield (node.key, node.value)
      yield from self._inorder(node.right)


  def _postorder(self, node):
    if node is not None:
      yield from self._postorder(node.left)
      yield from self._postorder(node.right)
      yield (node.key, node.value)


```
